```
import matplotlib.pyplot as plt
import numpy as np
```

Os dados são gerados artificialmente da seguinte forma: $y=2x-5 +\epsilon$. Os valores de $x$ são amostrados de $10*[0,1)$ a partir de uma distribuição e $\epsilon$ de uma distribuição normal com média 0 e desvio padrão 1.

```
rng = np.random.RandomState(1)
x = 10 * rng.rand(50)
y = 2 * x - 5 + rng.normal(0,1, 50)
plt.scatter(x, y)
```

Aqui usamos a biblioteca [scikit-learn](https://scikit-learn.org/stable/) escrita em Python para criar e treinar o modelo de regressão.

```
from sklearn.linear_model import LinearRegression

model = LinearRegression()

#Os dados originalmente estão no formato (nr. entradas,) e precisam estar como (nr. entradas, nr. atributos). O comando np.newaxis do numpy corrige o problema
model.fit(x[:,np.newaxis],y)

#Aqui geramos 1000 dados adicionais de 1 a 10 com espaçamento uniforme para gerar uma linha suave sobreposta aos dados de treino.
xfit = np.linspace(1,10,1000)
yfit = model.predict(xfit[:,np.newaxis])

plt.scatter(x,y)
plt.plot(xfit,yfit)

```

```
model.coef_
```

```
def plot_residuals_vs_fit(X,Y, model):
#    plot_data(X,Y)
#    x_plot = np.array([i/200.0 for i in range(200)])
#    X_plot = x_plot[:,np.newaxis]
    y_pred = model.predict(X)
    res = Y - y_pred
    plt.plot(y_pred,res,'k.',color='orange',)
    plt.axhline(y=0., color='r', linestyle='-')
    plt.xlabel("predictions")
    plt.ylabel("residuals")
```

```
plot_residuals_vs_fit(x[:,np.newaxis],y, model)
```
