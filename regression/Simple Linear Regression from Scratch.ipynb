```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 100            # Number of data samples
X = np.random.uniform(-5.0, 5.0, n)

mu, sigma = 0, 2     # Mean and standard deviation
noise = np.random.normal(mu, sigma, n)

y = 3*X - 2 + noise

plt.figure(figsize=(8,6))
plt.plot(X, y, 'ob')
plt.ylabel('y', rotation = 0, fontsize = 14)
plt.xlabel('x', fontsize = 14)
plt.title('Scatter plot of data')
plt.grid()
plt.show()
```

```python
class LinearRegression:

  def __init__(self, learning_rate=0.0001, n_iterations=100):
    self.learning_rate = learning_rate
    self.n_iterations = n_iterations
    self.weights = None

  def fit(self, X, y):
    n_samples, n_features = X.shape

    costs = []

    self.weights = np.zeros(n_features)

    for i in range(self.n_iterations):
      y_pred = np.dot(X, self.weights)
      e = y - y_pred
      rss = np.dot(e, e)
      costs.append(rss)
      print("RSS {}".format(rss))

      d_weights = -2*np.dot(X.T, e)

      self.weights -= self.learning_rate * d_weights

    return costs

  def predict(self, X):
    y_pred = np.dot(X, self.weights)
    return y_pred
```

```python
from sklearn.datasets import make_regression
X, y = make_regression(n_samples=1000, n_features=1, noise=0.1,random_state=0)
X = np.column_stack((np.ones(X.shape),X))
```

```python
X.shape
```

```python
import time
start = time.time()

model = LinearRegression(0.0005)
J = model.fit(X,y)
end = time.time()
print(end - start)
```

```python
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
plt.plot(range(len(J)), J,  '-xb')
plt.title('Simple Linear Regression')
plt.xlabel('Iterations')
plt.ylabel('Cost function')
plt.xticks(range(0,model.n_iterations+1,2))
plt.grid()
```
