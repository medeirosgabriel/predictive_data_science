```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
class LinearRegression:

  def __init__(self, learning_rate=0.01, n_iterations=100):
    self.learning_rate = learning_rate
    self.n_iterations = n_iterations
    self.w0 = 0.
    self.w1 = 0.

  def fit(self, X, y):

    costs = []

    for i in range(self.n_iterations):
      rss = 0.
      d_w0 = 0.
      d_w1 = 0.

      for j in range(X.shape[0]):

        y_pred = self.w0 + self.w1*X[j]
        e = y[j] - y_pred
        rss += e**2

        d_w0 += -2*e
        d_w1 += -2*e*X[j]

      costs.append(rss)
      self.w0 -= self.learning_rate*d_w0
      self.w1 -= self.learning_rate*d_w1
      print("RSS: {}".format(rss))
    return costs

  def predict(self, x):
    y_pred = self.w0 + self.w1*x
    return y_pred
```

```python
# Função ímpar 2X+1

X = np.array([1,2,3,4,5])
y = np.array([3,5,7,9,11])
```

```python
model = LinearRegression(n_iterations=50, learning_rate=0.0001)
J = model.fit(X,y)
print("Parametros: w0={} e w1={}".format(model.w0, model.w1))
```

```python
model.predict(7)
```

```python
# Dados sintéticos comruído

n = 100

X = np.random.uniform(-5, 5, n)

mu, sigma = 0, 2
noise = np.random.normal(mu, sigma, n)

y = 3*X - 2 + noise
```

```python
plt.figure(figsize=(8,6))
plt.plot(X,y, 'ob')
plt.ylabel('y', rotation=0, fontsize=14)
plt.xlabel('x', fontsize=14)
plt.title('Scatter plot of data')
```

```python
import time

start = time.time()
model = LinearRegression(learning_rate=0.0005)
J = model.fit(X,y)
end = time.time()
print(end - start)
print("Parametros: w0={} e w1={}".format(model.w0, model.w1))
```

```python
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
plt.plot(range(len(J)), J,  '-xb')
plt.title('Simple Linear Regression')
plt.xlabel('Iterations')
plt.ylabel('Cost function')
plt.xticks(range(0,model.n_iterations+1,2))
plt.grid()


y_pred = model.w1 * X + model.w0
#y_pred = obj.predict(x, True)

plt.subplot(1,2,2)
plt.plot(X, y, 'ob', label = "Actual")
plt.plot(X, y_pred, 'r', label = "Predicted")
plt.ylabel('y', fontsize= 14, rotation=0)
plt.xlabel('x', fontsize= 14)
plt.legend(loc='best')
plt.title('Simple Linear Regression')
plt.grid()
plt.show()
```
