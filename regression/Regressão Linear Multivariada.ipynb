```python
import numpy as np
a = np.random.uniform(-5.0, 5.0, 10000)
b = np.random.uniform(-5.0, 5.0, 10000)
```

#Com DOT

```python
import time
start = time.time()
c = np.dot(a,b)
end = time.time()
print(end - start)
c
```

#Com la√ßo

```python
start = time.time()
sum = 0
for i in range(len(a)):
  sum += a[i] * b[i]
end = time.time()
print(end - start)
sum
```
